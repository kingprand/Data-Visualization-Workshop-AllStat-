## Cross Tabulation
Analisis deskriptif yang digunakan untuk mengetahui hubungan dari dua variabel atau lebih. Analisis crosstab memungkinkan pengguna menyilangkan data pada variabel satu dengan variabel lainnya.

### Contoh

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['infantry', 'infantry', 'cavalry', 'cavalry', 'infantry', 'infantry', 'cavalry', 'cavalry','infantry', 'infantry', 'cavalry', 'cavalry'], 
        'experience': ['veteran', 'rookie', 'veteran', 'rookie', 'veteran', 'rookie', 'veteran', 'rookie','veteran', 'rookie', 'veteran', 'rookie'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['regiment', 'company', 'experience', 'name', 'preTestScore', 'postTestScore'])
df
```

```python
pd.crosstab(df["regiment"], df['company'])
```

Diatas merupakan hasil cross tabulation dari data df. Tujuan crosstab adalah untuk upaya meringkas data menjadi bentuk tabulasi silang. 

```python
pd.crosstab(df["regiment"], df['company'],margins=True)
```

Margin pada fungsi crosstab di pandas memiliki tujuan menampilkan total dari hasil tabulasi silang seperti terlihat pada hasil output diatas.

### Penjualan berdasarkan merk tiap lokasi pembelian jenis kelamin dan jenis pembayaran

> Sebelum melakukan visualisasi pada data bertipe string, kita harus memanipulasi data tersebut agar dapat divisualisasikan

```python
pd.crosstab(data['Jenis Kelamin'],data['Merek']).plot(kind='bar')
```

Dari hasil output diatas didapatkan bahwa penjualan semua merk hp lebih banyak laki-laki dibandingkan perempuan. Artinya laki-laki memiliki keinginan memiliki hp lebih tinggi dibanding perempuan. Namun merk hp favorit tiap jenis kelamin berbeda, laki-laki lebih banyak yang membeli samsung sedangkan perempuan lebih banyak membeli xiaomi.

```python
pd.crosstab(data['Jenis Pembayaran'],data['Merek']).plot(kind='bar')
```

Dari hasil output diatas didapatkan bahwa penjualan semua merk hp lebih banyak melakukan pembayaran menggunakan kredit dibandingkan tunai. Namun merk hp favorit tiap jenis pembayaran berbeda, pada pembayaran kredit lebih banyak digunakan untuk produk xiaomi sedangkan tunai lebih banyak digunakan untuk produk samsung.
